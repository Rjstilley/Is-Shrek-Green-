```python
import pandas as pd
```

```python
table = pd.read_csv("data/full_df.csv")
table.head()
```

```python
table = table[['ID','Patient Age','Patient Sex','Left-Diagnostic Keywords','Right-Diagnostic Keywords','N','D','G','C','A','H','M','O']]
table
```

```python
table = table.rename(columns={"N":"Normal","D":"Diabetes","G":"Glaucoma","C":"Cataract","A":"Age Related Macular Degeneration","H":"Hypertension","M":"Myopia","O":"Other"})
```

```python
table = table.drop_duplicates(subset=['ID'])
```

```python
table = table.set_index("ID")
```

```python
table.to_csv("data/table.csv",index=True,header=True)
```

```python
from sqlalchemy import create_engine
rds_connection_string = "postgres:postgres@localhost:5432/Eye"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
table.to_sql(name='Eye', con=engine, if_exists='append', index=True)
```
